```python
# Part 2 â€“ Visualization of Data using matplotlib, pyplots Packages

'''1. Generate a Data set for Health Care using the following: 
people = ['kiran', 'arun', 'vijay', 'varun']
age =[25, 30, 35, 40, 45]
height =[145, 151, 165, 173]
weight=[45, 55, 65, 75]'''

import pandas as pd
import matplotlib.pyplot as plt

people = ['kiran', 'arun', 'vijay', 'varun']
age =[25, 30, 35, 40]
height =[145, 151, 165, 173]
weight=[45, 55, 65, 75]

df = pd.DataFrame({'Name': people,'Age':age,'Height':height,'Weight':weight})

print(df)
```

```python
'''2. Using the generated dataset given above. 
Generate the following graphs and justify the relationships among the vectors.'''

# 3. Scatter Plot

plt.scatter(df['Height'], df['Weight'])
plt.title('Scatter Plot: Height vs Weight')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

```python
# 4. Bar Graph
plt.bar(df['Height'],df['Weight'])
plt.title('Bar Graph: Height vs Weight')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

```python
# 5. Histogram
plt.hist(df['Height'], bins=10, edgecolor='black')
plt.title('Histogram: Height')
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.show()
```
